```python
#part 1 assume uniform priors 
#site A click 30 visit 8000
#site B clicks 4 , visit 515 
# which site is better A or B
```

```python
import numpy as np
import statsmodels.api as sm
import pystan
from scipy.stats import uniform
# model for A, fed with the right values
sample_size=10000
views_A=8000
clicks_A=30
views_B=515
clicks_B=4
A_sample =uniform.rvs(1 + clicks_A,1 + views_A - clicks_A,size=sample_size)
# model for B, fed with the right values
B_sample =uniform.rvs(1 + clicks_B,1 + views_B - clicks_B,size=sample_size)
# let's find out the probability that A is better than B
print(np.mean(A_sample > B_sample))
# we can also find the probability that p_A is larger than p_B by 0
```

```python
print (np.mean(A_sample > (B_sample + 0.05)))
```

```python
plt.plot(B_sample)
plt.show()
```

```python
np.mean(A_sample)
```

```python
np.mean(B_sample)
```

```python
#A/B testing shows that Site A is doing better do to having abtter sample mean. 
```
